```python
import numpy as np
import random
```

```python
class Board:
  def __init__(self, size=10) -> None:
      self.size = size
      self.board = [['O' for _ in range(size)] for _ in range(size)]
      self.ships = []
      self.occupied = set()

  def place_ship(self, ship_sizes=[2, 3, 3, 4], seed=None) -> None:
    rng = random.Random(seed)
    for ship in ship_sizes:
        placed = False
        while not placed:
            horizontal = rng.choice([True, False])
            if horizontal:
                r = rng.randrange(self.size)
                c = rng.randrange(self.size - ship + 1)
                ship_cells = {(r, c + i) for i in range(ship)}
            else:
                r = rng.randrange(self.size - ship + 1)
                c = rng.randrange(self.size)
                ship_cells = {(r + i, c) for i in range(ship)}

            # 检查是否重叠
            if ship_cells.isdisjoint(self.occupied):
                self.ships.append(ship_cells)
                self.occupied |= ship_cells
                placed = True
```

```python
# test
c = Board()
c.place_ship()
print(c.ships)
```
